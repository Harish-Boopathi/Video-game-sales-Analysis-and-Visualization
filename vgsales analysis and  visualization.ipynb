```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(color_codes=True)
```

```python
df = pd.read_csv('C:/Users/haris/Documents/vgsales.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df1 = df.dropna()
df1
```

```python
dftotal = df1.groupby('Genre', as_index=False)['Global_Sales'].sum()
dftotal
```

```python
dftotal.nlargest(10,'Global_Sales')
```

```python
#Show Bar Chart
plt.figure(figsize=(10,8))
sns.barplot(data=dftotal, x='Global_Sales', y='Genre')
plt.title('Total Global Sales Each Genre', fontsize=18)
plt.xlabel ('Total Global Sales', fontsize=16)
plt.ylabel ('Genre', fontsize=16)
plt.show()
```

```python
dfyear=df1.groupby('Year',as_index=False)['Global_Sales'].sum()
dfyear
```

```python
dftopyear=dfyear.nlargest(5,'Global_Sales')
dftopyear
```

```python
plt.figure(figsize=(5,3))
sns.barplot(data=dftopyear,x='Year',y='Global_Sales')
plt.title("top year's with global sales")
plt.xlabel('Year')
plt.ylabel('Global sales')
plt.show()
```

```python
dfplatform = df1.groupby('Platform', as_index=False)['Global_Sales'].sum()
dfplatform
```

```python
dfplatform.nlargest(31,'Global_Sales')
```

```python
#Show Bar Chart
plt.figure(figsize=(10,8))
sns.barplot(data=dfplatform, x='Global_Sales', y='Platform')
plt.title('Total Global Sales Each Platform', fontsize=18)
plt.xlabel ('Total Global Sales', fontsize=16)
plt.ylabel ('Platform', fontsize=16)
plt.show()
```

```python
dfPublisher = df1.groupby('Publisher', as_index=False)['Global_Sales'].sum()
dfPublisher
```

```python
dfPublisher=dfPublisher.nlargest(10,'Global_Sales')
dfPublisher
```

```python
y = np.array(dfPublisher['Global_Sales'])
mylabels = dfPublisher['Publisher']

plt.pie(y, labels = mylabels,autopct='%1.2f%%')
plt.title('Total Global Sales Each Publisher', fontsize=18)
plt.show() 
```
